## Bar chart

```python
fruits = {"Apple": 5, "Orange": 10, "Strawberry":8, "Kiwi":15}

fig, ax = plt.subplots()
ax.bar(fruits.keys(), fruits.values())
ax.set(title="Fruit Market",
       xlabel="Fruit",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()

ax.barh(fruits.keys(), fruits.values())
```

## Histogram

```python
x = np.random.randn(1000) # normal distribution
fig, ax = plt.subplots()
ax.hist(x)
ax.set(title="Histogram",
       xlabel="Number",
       ylabel="Frequency");
```

## Subplots

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

x = [1, 2, 3, 4]
y = [100, 300, 150, 500]

ax1.plot(x, y)
ax2.scatter(np.random.rand(10), np.random.rand(10))
ax3.bar(fruits.keys(), fruits.values())
ax4.hist(np.random.randn(100));
```

```python
car_sales = pd.read_csv("data/car-sales.csv")
car_sales
```

```python
car_sales.Price = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True).str[:-2].astype(int)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2024", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sale"] = car_sales.Price.cumsum()
car_sales
```

```python
car_sales.plot("Sale Date", "Total Sale")
```

```python
car_sales.plot("Odometer (KM)", "Price", kind="scatter")
```

```python
car_sales.plot("Make", "Price", kind="bar")
```

```python
car_sales["Price"].plot(kind="hist")
```

```python
car_sales.plot(kind="hist", subplots=True)
```

```python
car_sales
```

```python
## OO methods
fig, ax = plt.subplots(figsize=(12,8))

## Plot data
scatter = ax.scatter(x=car_sales.Make,
                     y=car_sales.Price,
                     c=car_sales.Doors,
                     cmap="winter")

## Customize
ax.set_xlim()
ax.set_ylim([1000, 30000])

## Add Legend
ax.legend(*scatter.legend_elements(), title="Doors")

## Add line
ax.axhline(y=car_sales.Price.mean(), linestyle="--")
```
