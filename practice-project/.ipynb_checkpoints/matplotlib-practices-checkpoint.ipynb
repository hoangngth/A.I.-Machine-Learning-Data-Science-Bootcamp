```python
## 1. Prepare data
x = [1, 2, 3, 4]
y = [89, 34, 64, 45]

## 2. Setup plot
fig, ax = plt.subplots(figsize=(12,8))

## 3. Plot data
ax.plot(x,y)

## 4. Customize plot
ax.set(title="Example plot 1",
       xlabel="Label x",
       ylabel="Label y")

## 5. Save plot
fig.savefig("images/example-plot-1.png")
```

```python
x = np.linspace(0, 10 , 100)
x
```

```python
fruits = {"Apple": 5, "Orange": 10, "Strawberry":8, "Kiwi":15}

fig, ax = plt.subplots()
ax.bar(fruits.keys(), fruits.values())
ax.set(title="Fruit Market",
       xlabel="Fruit",
       ylabel="Price ($)");
```
